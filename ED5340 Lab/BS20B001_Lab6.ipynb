## Question 1

```python
import numpy as np
```

```python
class Matrix:
    def __init__(self,m1=np.zeros(0)):
        self.mat = np.array(m1)
    def elementwise_addition(self,m):
        if(self.mat.shape == m.mat.shape):
            return np.add(self.mat, m.mat)
        else:
            print('ValueError: shape of both matrices are different')
    def elementwise_subtraction(self,m):
        if(self.mat.shape == m.mat.shape):
            return np.subtract(self.mat,m.mat)
        else:
            print('ValueError: shape of both matrices are different')
    def elementwise_multiplication(self,m):
        if(self.mat.shape == m.mat.shape): 
            return np.multiply(self.mat,m.mat)
        else:
            print('ValueError: shape of both matrices are different')
    def scalar_multiplication(self,scalar):
        return np.array(self.mat)*scalar
    def determinant(self):
        if(self.mat.shape[0] == self.mat.shape[1]):
            det = np.linalg.det(self.mat)
            return det
        else:
            print('ValueError: Given matrix is not a square matrix')
    def matrix_multiplication(self,m):
        if(self.mat.shape[1] == m.mat.shape[0]):
            return np.matmul(self.mat, m.mat) 
        else:
            print('ValueError: columns of mat 1 is not matching rows of mat2')
    def transpose(self):
        return self.mat.T
    def inverse(self):
        if(self.mat.shape[0] != self.mat.shape[1]):
            print('ValueError: Given matrix is non invertible')
        elif(self.determinant() == 0):
            print('ValueError: Given matrix is non invertible')
        else:
            return np.linalg.inv(self.mat)
    def diagonal(self):
        return self.mat.diagonal()
    def rank(self):
        return np.linalg.matrix_rank(self.mat)
    def eigen(self):
        import matplotlib.pyplot as plt
        eig_val, eig_vec = np.linalg.eig(self.mat)
        if(eig_vec.shape[1] == 2):
            v1 = eig_vec[:, 0]
            v2 = eig_vec[:, 1]
            origin = [0,0]
            plt.quiver(*origin,v1[0],v1[1],color = ['b'], scale=5)
            plt.quiver(*origin,v2[0],v2[1],color = ['r'], scale=5)
            plt.xlim(-3,3)
            plt.ylim(-3,3)
            plt.show()
        else:
            v1 = eig_vec[:, 0]
            v2 = eig_vec[:, 1]
            v3 = eig_vec[:, 2]
            fig = plt.figure()
            ax = fig.gca(projection = '3d')
            origin = [0,0,0]
            ax.quiver(*origin,v1[0],v1[1],v1[2],color = ['b'], arrow_length_ratio = 0.2)
            ax.quiver(*origin,v2[0],v2[1],v2[2],color = ['r'], arrow_length_ratio = 0.2)
            ax.quiver(*origin,v1[0],v3[1],v3[2],color = ['g'], arrow_length_ratio = 0.2)
            ax.set_xlim([-1,1])
            ax.set_ylim([-1,1])
            ax.set_zlim([-1,1])
            plt.show()
        return eig_val, eig_vec
```

```python
m = Matrix([[1,2,4],[2,3,5],[4,5,3]])
m1 = Matrix([[2,3,9],[3,4,8],[5,6,5]])
print('elementwise addition:',m.elementwise_addition(m1))
print('elementwise subtraction:',m.elementwise_subtraction(m1))
print('elementwise multiplication:',m.elementwise_multiplication(m1))
print('scalar multiplication:',m.scalar_multiplication(3))
print('Det of m:',m.determinant())
print('Transpose of m: ',m.transpose())
print('Inverse of m: ',m.inverse())
print('Diagonal elements of m:', m.diagonal())
print('Rank of m: ',m.rank())
print('eigen value and eigen vector of m: ',m.eigen())
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = X**2 - Y**2
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.set_title('z = x^2 - y^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = np.sin(np.sqrt(X**2+Y**2))
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.set_title('z = sin((x^2 + y^2)^0.5)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = np.cos(X) + np.sin(Y)
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.set_title('z = cos(x)+sin(y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X,Y,Z)
plt.show()
```

```python
f = open('pledge.txt', 'w+')
f.write('India is my country and all Indians are my brothers and sisters. I love my country and I am proud of its rich and varied heritage. I shall always strive to be worthy of it. I shall give respect to my parents, teachers and elders and treat everyone with courtesy. To my country and my people, I pledge my devotion. In their well being and prosperity alone, lies my happiness. ')
dic = {}
for line in f:
    line = line.strip()
    words = line.split(' ')
    for word in words:
        if word in dic:
            dic[word]+=1
        else:
            dic[word] = 1
f1 = open('count.txt', 'w')
for i,j in dic.items():
    f1.write(i)
    f1.write(':')
    f1.write(str(j))
    f1.write('\n')
f1.close()
f.close()
```
