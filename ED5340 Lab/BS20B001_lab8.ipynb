```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def J(w1,w2):
    return ((w1-10)**2 + (w2-10)**2)
```

```python
w1 = np.linspace(-100,100)
w2 = np.linspace(-100,100)
W1,W2 = np.meshgrid(w1,w2)
J = J(W1,W2)
fig,ax = plt.subplots()
cs = ax.contour(W1,W2,J)
ax.clabel(cs, inline = 1)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(W1,W2,J)
plt.show()
```

```python
import numpy as np

def func(w):
  r = np.array([10, 10])
  return np.sum(np.square(w - r))

def midpt(a, b):
  return (a+b)/2

def line_search(str_pt, lst_pt, thrs):
  a = str_pt
  b = lst_pt
  trj = []

  while np.sqrt(np.sum(np.square(a - b))) > thrs:
    wm = (a + b)/2
    w1 = (a+wm)/2
    w2 = (wm + b)/2

    if func(w1) < func(w2):
      b = wm
      wmin = w1
    else:
      a = wm
      wmin = w2
    trj.append(wmin)

  return wmin, trj
```

```python
wimin, trj = line_search(np.array([2,5]),np.array([3,7]),0.0001)
print('Minimum : ',wimin)
```

```python
fig,ax = plt.subplots()
cs = ax.contour(W1,W2,J)
ax.clabel(cs, inline = 1)
for i in trj:
    ax.scatter(i[0],i[1])
    print(i[0],i[1])
plt.show()
```
